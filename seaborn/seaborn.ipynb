```python
#barplot

res = sns.barplot(mtcars['cyl'],mtcars['mpg'])
plt.show()
```

```python
res=sns.barplot(x='cyl',y='mpg',data=mtcars)
plt.show()
```

```python
res=sns.barplot(x='cyl',y='mpg',data=mtcars,color='yellow')
plt.show()
```

```python
res=sns.barplot(x='am',y='mpg',data=mtcars,palette="Set2")
plt.show()
```

```python
res=sns.barplot(x='gear',y='hp',data=mtcars,palette="rocket")
plt.show()
```

```python
#count plot
sns.countplot(x='cyl',data=mtcars,palette="Set1")
```

```python
sns.countplot(x='carb',data=mtcars,palette="Set2")
```

```python
sns.countplot(x='gear',hue='cyl',data=mtcars,palette='Set1') #hue= nos of bars in graph
```
